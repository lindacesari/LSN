```python
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.stats import norm , cauchy
import matplotlib.mlab as mlab
from scipy.optimize import curve_fit

%matplotlib inline

# nicer figures
import matplotlib as mpl
mpl.rcParams['axes.titlesize'] = 16
mpl.rcParams['axes.labelsize'] = 14
mpl.rcParams['legend.fontsize'] = 12
mpl.rcParams['xtick.labelsize'] = 12
mpl.rcParams['ytick.labelsize'] = 12
mpl.rcParams['axes.spines.right'] = False
mpl.rcParams['axes.spines.top'] = False
plt.style.use('dark_background')

k_b = 1.380649 * 10**(-23) #(J/K)
epsilon = 120 * k_b  #(joule)
sigma = 0.34 * 10**(-9) #(m)
m = 6.634 * 10**(-26) #(kg)
```

```python
e_kin  = np.loadtxt( "MolecularDynamics_NVE/equilibration/solid/output_ekin.dat" , dtype = float , delimiter = "," , usecols=(0) , unpack='true')
e_pot  = np.loadtxt( "MolecularDynamics_NVE/equilibration/solid/output_epot.dat" , dtype = float , delimiter = "," , usecols=(0) , unpack='true')
e_tot  = np.loadtxt( "MolecularDynamics_NVE/equilibration/solid/output_etot.dat" , dtype = float , delimiter = "," , usecols=(0) , unpack='true')
temp  = np.loadtxt( "MolecularDynamics_NVE/equilibration/solid/output_temp.dat" , dtype = float , delimiter = "," , usecols=(0) , unpack='true')

#SI UNITS
e_kin *= epsilon 
e_pot *= epsilon
e_tot *= epsilon
temp *= 120 
```

```python
fig, ax = plt.subplots(1,1 , figsize=(13,8))
ax.plot(np.arange(len(e_kin)) , temp , color="#ECA5E8" , linewidth=3 , label = "Temperature")
ax.hlines(0.8*120 , 0 , 6000 , color = "White" , linewidth = 3 , label ="Expected temperature")
ax.set_title("Equilibration of a solid system" , fontsize=25)
ax.set_xlabel('Time steps')

ax.grid(True)
ax.legend(loc="best")
```

```python
fig, ax = plt.subplots(1,1 , figsize=(13,8))
ax.plot(np.arange(len(e_kin)) , e_kin , color = "#F3C16F", linewidth=3, label = "Kinetic energy")
ax.plot(np.arange(len(e_kin)) , e_pot , color = "#CCCCFF" , linewidth=3 , label = "Potential energy")
ax.plot(np.arange(len(e_kin)) , e_tot , color = "#A5EDB5" , linewidth=3 , label = "Total energy")

ax.set_title("Equilibration of a solid system" , fontsize=25)
ax.set_xlabel('Time steps')

ax.grid(True)
ax.legend(loc="best")
```

## SOLID PHASE

```python
eksolid , errorksol = np.loadtxt( "MolecularDynamics_NVE/solid/ave_ekin.dat" , dtype = float , delimiter = "," , usecols=(0,1) , unpack='true')
epotsol , errorpsol = np.loadtxt( "MolecularDynamics_NVE/solid/ave_epot.dat" , dtype = float , delimiter = "," , usecols=(0,1) , unpack='true')
etotsol , errortsol = np.loadtxt( "MolecularDynamics_NVE/solid/ave_etot.dat" , dtype = float , delimiter = "," , usecols=(0,1) , unpack='true')
tempsol , errortempsol = np.loadtxt( "MolecularDynamics_NVE/solid/ave_temp.dat" , dtype = float , delimiter = "," , usecols=(0,1) , unpack='true')

#SI UNITS
eksolid *= epsilon 
epotsol *= epsilon
etotsol *= epsilon
tempsol *= 120 
errorksol *= epsilon
errorpsol *= epsilon
errortsol *= epsilon
errortempsol *= 120
```

```python
fig, ax = plt.subplots(1,1 , figsize=(10,8))
ax.errorbar(np.arange(len(eksolid)) , eksolid , yerr=errorksol , color = "#F3C16F", linewidth=3, label = "Kinetic energy")
ax.errorbar(np.arange(len(eksolid)) , epotsol , yerr=errorpsol , color = "#CCCCFF" , linewidth=3 , label = "Potential energy")
ax.errorbar(np.arange(len(eksolid)) , etotsol , yerr=errortsol , color = "#A5EDB5" , linewidth=3 , label = "Total energy")
ax.set_title("Simulation of a solid" , fontsize=20)
ax.set_xlabel('Blocks')
ax.set_ylabel('Energy (Joule)')
ax.grid(True)
ax.legend(loc="best")
```

## LIQUID PHASE

```python
ekliq , errorkliq = np.loadtxt( "MolecularDynamics_NVE/liquid/ave_ekin.dat" , dtype = float , delimiter = "," , usecols=(0,1) , unpack='true')
epotliq , errorpliq = np.loadtxt( "MolecularDynamics_NVE/liquid/ave_epot.dat" , dtype = float , delimiter = "," , usecols=(0,1) , unpack='true')
etotliq , errortliq = np.loadtxt( "MolecularDynamics_NVE/liquid/ave_etot.dat" , dtype = float , delimiter = "," , usecols=(0,1) , unpack='true')
templiq , errortempliq = np.loadtxt( "MolecularDynamics_NVE/liquid/ave_temp.dat" , dtype = float , delimiter = "," , usecols=(0,1) , unpack='true')

#SI UNITS
ekliq *= epsilon 
epotliq *= epsilon
etotliq *= epsilon
templiq *= 120 
errorkliq *= epsilon
errorpliq *= epsilon
errortliq *= epsilon
errortempliq *= 120
```

```python
fig, ax = plt.subplots(1,1 , figsize=(10,8))
ax.errorbar(np.arange(len(ekliq)) , ekliq , yerr=errorkliq*10 , color = "#F3C16F", linewidth=3, label = "Kinetic energy")
ax.errorbar(np.arange(len(ekliq)) , epotliq , yerr=errorpliq*10 , color = "#CCCCFF" , linewidth=3 , label = "Potential energy")
ax.errorbar(np.arange(len(ekliq)) , etotliq , yerr=errortliq*100 , color = "#A5EDB5" , linewidth=3 , label = "Total energy")
ax.set_title("Simulation of a liquid system" , fontsize=20)
ax.set_xlabel('Blocks')
ax.set_ylabel('Energy (Joule)')
ax.grid(True)
ax.legend(loc="best")
```

```python
fig, ax = plt.subplots(1,1 , figsize=(10,8))
ax.errorbar(np.arange(len(ekliq)) , templiq , yerr=errortempliq , color="#ECA5E8" , linewidth=3 , label = "Temperature")
ax.hlines(1.1*120 , 0 , 99 , color = "White" , linewidth = 3 , label ="Expected temperature")
ax.set_title("Temperature in the simulation of a liquid system" , fontsize=20)
ax.set_xlabel('Blocks')
ax.set_ylabel('Temperature (K)')
ax.grid(True)
ax.legend(loc="best")
```
