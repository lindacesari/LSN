## <font color="fuxia"> direct sampling

```python
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy import stats
from scipy import optimize
from scipy.stats import norm
```

```python
# reading data from direct.csv

sum_prog, err_prog = np.genfromtxt("1/direct.csv", dtype=float, delimiter=',', unpack=True)
```

```python
import scipy
from scipy.special import erf
from scipy import log, exp, sqrt

# data
S0 = 100.
K = 100.
T = 1.
r = 0.1
sigma = 0.25
t=0

def N(x):
    return 0.5 * (1. + erf(x / np.lib.scimath.sqrt(2.)))

def black_scholes(S0, K, T, r, sigma):
    d1 = 1./(sigma * np.lib.scimath.sqrt(T)) * (np.lib.scimath.log(S0 / K) + (r + (sigma**2) / 2.) * T)
    d2 = d1 - sigma * np.lib.scimath.sqrt(T)
    C = S0 * N(d1) - K * np.exp(-r * T) * N(d2)
    P = S0 *(N(d1) - 1.) - K * np.exp(-r * T) * (N(d2)-1.)
    return C, P

call, put = black_scholes(S0, K, T, r, sigma)
```

```python
# plot progressive data

M = 100000
N = 100
L = int(M/N)
x = [ i*L for i in range(1, N+1) ]  # progressive throws

delimiter = N

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

ax1.errorbar(x, sum_prog[:N], yerr=err_prog[:N], color='dodgerblue', label='computed')
ax1.set_title("Direct sampling - CALL")
ax1.set_xlabel("number of throws")
ax1.set_ylabel(r"< C [S(0), 0] >")
ax1.hlines(call, 0, M, color='navy', label=f'expected ='+str(call))
ax1.grid(True)
ax1.legend()

ax2.errorbar(x, sum_prog[N:], yerr=err_prog[N:], color='salmon', label='computed')
ax2.set_title("Direct sampling - PUT")
ax2.set_xlabel("number of throws")
ax2.set_ylabel(r"< P [S(0), 0] >")
ax2.hlines(put, 0, M, color='red', label=f'expected ='+str(put))
ax2.grid(True)
ax2.legend()
```
