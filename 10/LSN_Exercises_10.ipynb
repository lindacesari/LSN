<span style="color:purple">

The main difference between the genetic algorithm of exercise 9 and the simulated annealing algorithm is that the latter works on a single chromosome and tries to make that one better and better instead of working with the population altogether. 
    
The simulated anneling is based on a slow cooling down to reach the best configuration, therefore the simulation was performed at different temperatures shown in the plots below. For every temperature 10000 moves were made and results were recorded every 10 moves.
    
A move of the simulated annealing algorithm consists in trying a mutation on the chromosome and computing its new fitness: if the fitness is better than before, the move is always accepted; otherwise, the move is accepted with a Boltzmann probability $ e^{-\frac{fit_{new}-fit_{old}}{T}}$, which means at low temperatures only few bad moves are accepted and the system goes towards the best configuration.
    
The possible mutations are the same as in exercise 9:
1. Swap of two random cities in the chromosome
2. Swap of two random adjacent cities in the chromosome
3. Shift of part of a chromosome
4. Group permutation of a random number of adjacent cities
5. Inversion of a random sized group of adjacent cities
    
The probability of mutation was set to 50%.
    
There is no crossover.
    
Like in exercise 9, to compute the fitness of the chromosome, $L^{(1)}$ was used.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

from matplotlib.pyplot import text
```

```python
steps = 1e4
n_print = 10
temps = np.genfromtxt("1/temperatures.csv")
colors = ['red', 'orangered', 'coral', 'orange', 'gold', 'yellow', 'greenyellow', 'limegreen', 'mediumspringgreen', 'aquamarine', 'lightskyblue', 'dodgerblue', 'blue', 'navy']
```

### <span style="color:fuchsia">  Circle</span>

```python
# get position of the 32 cities in a square

cities = 32

positions = np.genfromtxt("1/pos_circle.csv", delimiter=',', dtype=float, skip_header=1)
pos_x = positions[:,0]
pos_y = positions[:,1]
```

```python
# plot evolution of length of the best path (fitness of the first cromo in the ordered population) 
#   as a function of #generations computed

fig, ax = plt.subplots(1, 1, figsize=(15,5))

cost = np.genfromtxt("1/cost_circle.csv", dtype=float, delimiter=',')

ax.plot(np.arange(len(cost))*n_print, cost, label='best path length', linewidth=0.5)
for i in range(len(temps)):
    ax.axvline(i*steps, color=colors[i], linestyle='dashed')
    text(i*steps-2000, 40, "T = "+str(temps[i]), rotation=90, verticalalignment='center', color=colors[i])
ax.set_xlabel('generation')
ax.set_ylabel('path length')
ax.set_title('Best path length')

print("\nStarting path length = ", cost[0])
print("Best path length = ", cost[-1], '\n')
```

```python
# plot the best path

paths = np.genfromtxt("1/best_cromo_circle.csv", dtype=int, delimiter=',')

fig, axx = plt.subplots(1, 2, figsize=(20,8))
    
for ind in [0,1]:
    ax = axx[ind]
    ax.scatter(x=pos_x, y=pos_y, color='red', marker='*')
    for i in range(cities):
        ax.annotate(str(i+1), positions[i])
    path = paths[ind]
    for i in range(cities):
        x1 = pos_x[path[i-cities*int(i/cities)]-1]
        y1 = pos_y[path[i-cities*int(i/cities)]-1]
        x2 = pos_x[path[i+1-cities*int((i+1)/cities)]-1]
        y2 = pos_y[path[i+1-cities*int((i+1)/cities)]-1]
        ax.plot([x1, x2], [y1, y2], color='pink')
    if ind == 0:
        title = 'Starting path'
    else:
        title = 'Optimal path'
    ax.set_title(title)
    ax.set_ylabel('y')
    ax.set_xlabel('x')
    
print("\nStarting path =", paths[0])
print("Optimal path =", paths[1], '\n')
```

Two simulations of simulated annealing were performed, one with cities on a circumference and one with cities in a square, the same used in the previous exercise (using seed=1).

The parallelized genetic algorithm works like explained in notebook 9, except that there is some sharing of the best chromosomes. 
    
In particular the algorithm worked on 4 cores and every 25 generations the 10 best chromosomes of each process were shared amongst the 4 cores. The 40 resulting chromosomes were put at the end of the sorted population, substituting the worst ones.
    
The simulation was performed on a population of 300 chromosomes, each with 32 genes and the total number of generations was 1000 for each core. The probability of crossover was 60% and of mutation was 50%.
    
Obviously the 4 cores worked on the city scheme.

```python
# get position of the 32 cities in a square

cities = 32

positions = np.genfromtxt("2/pos_square.csv", delimiter=',', dtype=float, skip_header=1)
pos_x = positions[:,0]
pos_y = positions[:,1]
```

```python
# plot evolution of length of the best path (fitness of the first cromo in the ordered population) 
#   as a function of #generations computed

cores = [0, 1, 2, 3, 4]    # core = 4 is for the single core run

fig, ax = plt.subplots(1, 1, figsize=(10,6))
for core in cores:
    cost = np.genfromtxt("2/cost"+str(core)+".csv", dtype=float, delimiter=',')[:-1]
    if core == 4:
        label = "single core"
    else:
        label = f"core {core}"
    ax.plot(np.arange(len(cost)), cost, label=label)
    print("Best path length for",label,"=",cost[-1])
ax.set_xlabel('generation')
ax.set_ylabel('length of best path')
ax.set_title('Best path length')
ax.legend()
```

```python
# plot the best path found by each core

fig, axx = plt.subplots(len(cores), 1, figsize=(7,30))
    
for core in cores:
    ax = axx[core]
    ax.scatter(x=pos_x, y=pos_y, color='red', marker='*')
    for i in range(cities):
        ax.annotate(str(i+1), positions[i])
    opt_path = np.genfromtxt("2/best_cromo"+str(core)+".csv", dtype=int, delimiter=',')[-1]
    for i in range(cities):
        x1 = pos_x[opt_path[i-cities*int(i/cities)]-1]
        y1 = pos_y[opt_path[i-cities*int(i/cities)]-1]
        x2 = pos_x[opt_path[i+1-cities*int((i+1)/cities)]-1]
        y2 = pos_y[opt_path[i+1-cities*int((i+1)/cities)]-1]
        ax.plot([x1, x2], [y1, y2], color='pink')
    if core == 4:
        label = "Single core process"
    else:
        label = f"4 cores - process {core}"
    ax.set_title(label)
    ax.set_ylabel('y')
    ax.set_xlabel('x')
    print('\n',label, "\n\toptimal path =",opt_path)
```

<span style="color:purple">

The first plot shows the length of the best path as a function of the generations perfomed, both for each of the 4 cores and for the case run with a single core. It is clear that both the 4 cores and the single core reach the same best chromosome, but the 4 core case is much faster in doing so, needing less than 200 generations, while the single core needs about 500.
    
What we gain in using different connected continents is therefore rapidity of convergence.
    
The 5 plots below show the final best path and that is clearly a good path since there are no intersections. It is the same path for both the 4 cores and the single core, as we were able say by looking at the best path length plot.
